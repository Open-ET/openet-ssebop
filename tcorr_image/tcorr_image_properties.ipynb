```python
import pprint
import subprocess

import ee
ee.Initialize()
```

This block is for copying the old Tcorr daily images to the new version with the date_ingested in the image ID.

This block will also set all the properties to lowercase and rename a few of them.

```python
tcorr_img_coll_id = 'projects/usgs-ssebop/tcorr_image/topowx_median_v0_daily'
tcorr_img_coll = ee.ImageCollection(tcorr_img_coll_id)

start_date = '2019-01-01'
end_date = '2019-05-01'

tcorr_info = tcorr_img_coll.filterDate(start_date, end_date).getInfo()
tcorr_id_list = tcorr_img_coll.filterDate(start_date, end_date).aggregate_array('system:id').getInfo()
# pprint.pprint(tcorr_id_list)

for tcorr in tcorr_info['features']:
    # Skip Tcorr images that already have the export date in the ID
    if len(tcorr['id'].split('/')[-1]) > 8:
        print('{} - skipping'.format(tcorr['id']))
        continue
    print('{}'.format(tcorr['id']))
    
    # Create a copy of the image with the export date in the ID
    if 'EXPORT_DATE' in tcorr['properties']:
        copy_id = tcorr['id'] + '_' + tcorr['properties']['EXPORT_DATE'].replace('-', '')
    elif 'date_ingested' in tcorr['properties']:
        copy_id = tcorr['id'] + '_' + tcorr['properties']['date_ingested'].replace('-', '')
    else:
        # print('  no date_ingested or EXPORT_DATE, skipping')
        continue
    # print('  {}'.format(copy_id))
        
    if copy_id in tcorr_id_list:
        # print('  image already exists, skipping')
        continue
        
    # print('  copying image')
    ee.data.copyAsset(tcorr['id'], copy_id, False)
```

```python
tcorr_img_coll_id = 'projects/usgs-ssebop/tcorr_image/topowx_median_v0_daily'
tcorr_img_coll = ee.ImageCollection(tcorr_img_coll_id)

start_date = '2019-01-01'
end_date = '2019-03-01'

tcorr_info = tcorr_img_coll.filterDate(start_date, end_date).getInfo()
tcorr_id_list = tcorr_img_coll.filterDate(start_date, end_date).aggregate_array('system:id').getInfo()
# pprint.pprint(tcorr_id_list)

for tcorr in tcorr_info['features']:
    # Skip Tcorr images that already have the export date in the ID
    if len(tcorr['id'].split('/')[-1]) <= 8:
        print('{} - skipping'.format(tcorr['id']))
        continue
    print('{}'.format(tcorr['id']))
    
    # Modify the properties of the image copy
    # print('  setting properties')
    # pprint.pprint(tcorr['properties'])
    properties = {}
    if 'cycle_day' not in tcorr['properties'] and 'CYCLE_DAY' in tcorr['properties']:
        properties['cycle_day'] = tcorr['properties']['CYCLE_DAY']
        properties['CYCLE_DAY'] = None
    if 'date' not in tcorr['properties'] and 'DATE' in tcorr['properties']:
        properties['date'] = tcorr['properties']['DATE']
        properties['DATE'] = None
    if 'day' not in tcorr['properties'] and 'DAY' in tcorr['properties']:
        properties['day'] = tcorr['properties']['DAY']
        properties['DAY'] = None
    if 'doy' not in tcorr['properties'] and 'DOY' in tcorr['properties']:
        properties['doy'] = tcorr['properties']['DOY']
        properties['DOY'] = None
    if 'date_ingested' not in tcorr['properties'] and 'EXPORT_DATE' in tcorr['properties']:
        properties['date_ingested'] = tcorr['properties']['EXPORT_DATE']
        properties['EXPORT_DATE'] = None
    if 'landsat' not in tcorr['properties'] and 'LANDSAT' in tcorr['properties']:
        properties['landsat'] = tcorr['properties']['LANDSAT']
        properties['LANDSAT'] = None
    if 'model_version' not in tcorr['properties'] and 'SSEBOP_VERSION' in tcorr['properties']:
        properties['model_version'] = tcorr['properties']['SSEBOP_VERSION']
        properties['SSEBOP_VERSION'] = None 
    if 'model_name' not in tcorr['properties']:
        properties['model_name'] = 'SSEBOP'
    if 'month' not in tcorr['properties'] and 'MONTH' in tcorr['properties']:
        properties['month'] = tcorr['properties']['MONTH']
        properties['MONTH'] = None
    if 'tmax_source' not in tcorr['properties'] and 'TMAX_SOURCE' in tcorr['properties']:
        properties['tmax_source'] = tcorr['properties']['TMAX_SOURCE']
        properties['TMAX_SOURCE'] = None 
    if 'tmax_version' not in tcorr['properties'] and 'TMAX_VERSION' in tcorr['properties']:
        properties['tmax_version'] = tcorr['properties']['TMAX_VERSION']
        properties['TMAX_VERSION'] = None 
    if 'year' not in tcorr['properties'] and 'YEAR' in tcorr['properties']:
        properties['year'] = tcorr['properties']['YEAR']
        properties['YEAR'] = None
    if 'wrs2_tiles' not in tcorr['properties'] and 'WRS2_TILES' in tcorr['properties']:
        properties['wrs2_tiles'] = tcorr['properties']['WRS2_TILES']
        properties['WRS2_TILES'] = None
        
    # pprint.pprint(properties)

    if not properties:
        continue
        
    ee.data.setAssetProperties(copy_id, properties)
```
